```python
##loading the required libraries
from sklearn import tree
import msmbuilder as msb
import mdtraj as mdt
import numpy as np
%pylab inline
```

```python
##helper functions 
colorlist=['b', 'g', 'r', 'y']

w=lambda x: np.where(data[:,-1]==x)
w_cls=lambda x: np.where(np.array(y_test)==x)

def PlotIthState(i,func=w):
    if func!="w_cls":
        scatter(data[func(i),0],data[func(i),1],color=colorlist[i],label="State %d"%i)
        axis([-180,180,-180,180])
    else:
        scatter(np.array(x_test)[func(i),0],np.array(x_test)[func(i),1],color=colorlist[i],label="State %d"%i)
        axis([-180,180,-180,180])
        
```

```python
mac=mdt.io.loadh("MacroAssignments.h5")
NumStates= np.max(mac['Data'])+1
phi=mdt.io.loadh("Phi.h5")
psi=mdt.io.loadh("Psi.h5")
```

```python
n_traj = np.shape(mac['Data'])[0]
n_frames_per_traj = np.shape(mac['Data'])[1]
n_frames = n_traj*n_frames_per_traj
```

```python
##get the data and the noise
data=np.zeros((n_frames,11))

for i in range(100):
    for j in range(501):
        data[i*501+j,0]=phi['Data'][i,j]
        data[i*501+j,1]=psi['Data'][i,j]
        data[i*501+j,2]=np.random.normal()
        data[i*501+j,3]=np.random.normal()
        data[i*501+j,4]=np.random.normal()
        data[i*501+j,5]=np.random.normal()
        data[i*501+j,6]=np.random.normal()
        data[i*501+j,7]=np.random.normal()
        data[i*501+j,8]=np.random.normal()
        data[i*501+j,9]=np.random.normal()
        data[i*501+j,10]=mac['Data'][i,j]

        
data=np.random.permutation(data)
x_all=data[:,:10]
y_all=data[:,10]
#
```

```python
##plotting the oiriginal dataset
figure(figsize=(8,6))
for i in range(NumStates):
    print(i)
    PlotIthState(i,w)
title("Alanine Dipeptide Macrostates",fontsize=30)
xlabel(r"$\phi (\circ)$",fontsize=30)
ylabel(r"$\psi (\circ)$",fontsize=30)
axis([-180,180,-180,180])
plt.grid()
plt.box()
xticks(size=16)
yticks(size=16)
legend(loc='best')
tight_layout()
savefig("Original_ala.tiff",dpi=300)
```

## Getting 50% of conformations from all 4 states

```python
x_train=[]
y_train=[]
x_test=[]
y_test=[]
for i in np.arange(4):
    ind = np.where(y_all==i)
    my_len = np.int(0.5*np.shape(ind)[1])
    train_ind = ind[0][:my_len]
    test_ind = ind[0][my_len:]
    x_train.extend(x_all[train_ind])
    y_train.extend(y_all[train_ind])
    x_test.extend(x_all[test_ind])
    y_test.extend(y_all[test_ind])
```

## starting the Decision Tree analysis

```python
print np.shape(x_train)
```

```python
print "Max_Depth %Error"
for i in range(1,5):
    clf_gini = tree.DecisionTreeClassifier(max_depth=i,criterion="gini",max_features=None)
    clf_gini = clf_gini.fit(x_train,y_train)
    clf_gini_res=clf_gini.predict(x_test)

    print "%d\t"%(i),(1-float(np.sum(clf_gini_res==y_test))/len(y_test))*100
```
